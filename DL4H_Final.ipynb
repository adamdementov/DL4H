## Preprocess Lexicon

```python
english_words_fpath = 'cim-misspelling/data/english_words/words.txt'
umls_dir = 'cim-misspelling/data/umls/'
lexicon_out_dir = 'cim-misspelling/data/lexicon'
lexicon_out_fname = 'lexicon.json'

alpha_set = set("abcdefghijklmnopqrstuvwxyz")
allowed_char_set = set("abcdefghijklmnopqrstuvwxyz'-&")
char_set = set("0123456789abcdefghijklmnopqrstuvwxyz+-*/^.,;:=!?'()[]{} ")
```

```python
# Load english words
with open(english_words_fpath, 'r') as fd:
    english_vocab = [l.strip().lower() for l in fd.readlines()]
english_vocab = [w for w in english_vocab if (all([c in allowed_char_set for c in w]) and w[0] in alpha_set)]
english_vocab_set = set(english_vocab)
english_vocab = sorted(list(english_vocab_set))
print(f'{len(english_vocab)} English words loaded')
```

```python
# Load UMLS lexicons
lrwd_vocab_set = set()
with open(os.path.join(umls_dir, 'LRWD'), 'r') as fd:
    lines = [l.strip() for l in fd.readlines()]

for line in tqdm(lines):
    word, eui, temp = line.split('|')
    assert not temp
    if all([c in allowed_char_set for c in word]) and (word[0] in alpha_set):
        for splited in word.split():   # Actually all `word` s don't have a space.
            lrwd_vocab_set.add(splited.lower())
    
lrwd_vocab = sorted(list(lrwd_vocab_set))
print(f'LEXICON LRWD has {len(lrwd_vocab)} words')
```

```python
prevariants_vocab_set = set()
with open(os.path.join(umls_dir, 'prevariants'), 'r') as fd:
    lines = [l.strip() for l in fd.readlines()]

for line in tqdm(lines):
    phrase, eui, temp = line.split('|')
    for word in phrase.split():
        if all([c in allowed_char_set for c in word]) and (word[0] in alpha_set):
            prevariants_vocab_set.add(word.lower())
    
prevariants_vocab = sorted(list(prevariants_vocab_set))
print(f'LEXICON prevariants has {len(prevariants_vocab)} words')
```

```python
print(f'LEXICON has {len(lrwd_vocab_set | prevariants_vocab_set)} words')
```

```python
# Add all our vocabulary
total_vocab_set = lrwd_vocab_set | prevariants_vocab_set | english_vocab_set
total_vocab = sorted(list(total_vocab_set))
print(f'Total vocab has {len(total_vocab)} words')
```

```python
# Write the lexicon into a file
if not os.path.exists(lexicon_out_dir):
    print(f'Creating directory {lexicon_out_dir}')
    os.makedirs(lexicon_out_dir)
    
lexicon_out_fpath = os.path.join(lexicon_out_dir, lexicon_out_fname)
with open(lexicon_out_fpath, 'w') as fd:
    print(f'Writing lexicon to {lexicon_out_fpath}')
    json.dump(total_vocab, fd)
```

```python
df_notes = pd.read_csv(mimic_csv_fpath, low_memory=False)
print(f'{len(df_notes)} rows')
```

```python
def is_nan(obj):
    return isinstance(obj, float) and math.isnan(obj)

def find_rows(hadm_id, chartdate, charttime, storetime, category):
    df_temp = df_notes
    
    if is_nan(hadm_id):
        df_temp = df_temp[df_temp.HADM_ID.isnull()]
    else:
        df_temp = df_temp[df_temp.HADM_ID == hadm_id]
    if len(df_temp) == 1: return df_temp
    
    if is_nan(chartdate):
        df_temp = df_temp[df_temp.CHARTDATE.isnull()]
    else:
        df_temp = df_temp[df_temp.CHARTDATE == chartdate]
    if len(df_temp) == 1: return df_temp
    
    if is_nan(charttime):
        df_temp = df_temp[df_temp.CHARTTIME.isnull()]
    else:
        df_temp = df_temp[df_temp.CHARTTIME == charttime]
    if len(df_temp) == 1: return df_temp
    
    if is_nan(storetime):
        df_temp = df_temp[df_temp.STORETIME.isnull()]
    else:
        df_temp = df_temp[df_temp.STORETIME == storetime]
    if len(df_temp) == 1: return df_temp
    
    if is_nan(category):
        df_temp = df_temp[df_temp.CATEGORY.isnull()]
    else:
        df_temp = df_temp[df_temp.CATEGORY == category]
    if len(df_temp) == 1: return df_temp
    
    return df_temp
```

```python
row_ids, char_idxs = [], []

for i, annotation in tqdm(enumerate(clinspell_annotations), total=len(clinspell_annotations)):
    row_id, hadm_id, chartdate, charttime, storetime, category, old_char_idx, misspelling, correction = annotation
    rows = find_rows(hadm_id, chartdate, charttime, storetime, category)
    
    row, char_idx, min_dist = None, -1, max(rows.TEXT.map(len))
    for _, cand_row in rows.iterrows():
        text = cand_row.TEXT.lower()
        cand_idxs = [m.start() for m in re.finditer(misspelling, text)]
        for cand_idx in cand_idxs:
            dist = abs(old_char_idx - cand_idx)
            if dist < min_dist:
                row, char_idx, min_dist = cand_row, cand_idx, dist
    
    row_ids.append(row.ROW_ID)
    char_idxs.append(char_idx)
    assert row.TEXT[char_idx:char_idx+len(misspelling)].lower() == misspelling
```

```python
misspellings, corrections = list(zip(*clinspell_annotations))[-2:]
```

```python
def get_context_text(row_id, char_idx, misspelling):
    text = df_notes[df_notes.ROW_ID == row_id].iloc[0].TEXT
    def process_clinspell(s):
        s = re.sub(r'(?<!\n)\n(?!\n)', "\t", s)
        s = re.sub("\n\n", "\n", s)
        return s
    return process_clinspell(text[:char_idx]), process_clinspell(text[char_idx+len(misspelling):])
```

```python
contexts = []
for row_id, char_idx, misspelling in zip(row_ids, char_idxs, misspellings):
    context = get_context_text(row_id, char_idx, misspelling)
    contexts.append(context)
```
